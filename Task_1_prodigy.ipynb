```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for demonstration
# Replace this with your actual data
top_10_female_countries = pd.DataFrame({
      'Country': ['Monaco', 'Hong Kong', 'Macao', 'Japan', 'South Korea', 'Malta', 'French Polynesia', 'Guadeloupe', 'Martinique', 'Australia'],
    'Female Life Expectancy': [88.4, 88.3, 87.8, 87.7, 86.7, 86.2, 85.8, 85.8, 85.8, 85.7]
})

top_10_male_countries = pd.DataFrame({
    'Country': [ 'Australia','Monaco', 'Macao', 'Hong Kong', 'Japan', 'Norway', 'Iceland', 'Switzerland', 'New Zealand', 'Ireland'],
    'Male Life Expectancy': [84.8, 82.9, 82.6, 82.2, 81.6, 81.5, 81.1, 81.0, 80.9, 80.8]
})

# Plotting
plt.figure(figsize=(10, 6))

# Plot Female Life Expectancy
plt.barh(top_10_female_countries['Country'], top_10_female_countries['Female Life Expectancy'], color='skyblue', label='Female')

# Plot Male Life Expectancy
plt.barh(top_10_male_countries['Country'], top_10_male_countries['Male Life Expectancy'], color='salmon', label='Male')

# Add labels and title
plt.xlabel('Life Expectancy')
plt.ylabel('Country')
plt.title('Comparison of Male and Female Life Expectancy in the year 2020')
plt.legend()

# Show plot
plt.gca().invert_yaxis()  # Invert y-axis to have the highest value at the top
plt.show()
```


- The plotting is a **bar graph** comparing male and female life expectancy in 2020.
- It shows data for various countries including Monaco, Hong Kong, Macao, Japan, South Korea, Malta, French Polynesia, Guadeloupe, Martinique, Australia, Norway, Iceland, Switzerland, New Zealand, and Ireland.
